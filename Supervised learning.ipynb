# Simple Regression Model

```python
import pandas as pd
df = pd.read_csv("placement - placement.csv")
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("package")
#plt.show()
```

```python
df.iloc[:,0]
```

```python
df.iloc[:,-1]
```

```python
X= df[['cgpa']]
Y= df[['package']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.20,random_state=29)
```

```python
X_train
```

```python
X_test
```

```python
print(X_train.shape,X_test.shape)
```

```python
print(Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#training accuracy
model.score(X_train,Y_train)
```

```python
# this is the bulk prediction
Y_predict=model.predict(X_test)
```

```python
Y_test
```

```python
Y_predict
```

```python
#r2_score is used for numerical data and accuracy_score used for categorical data
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,Y_predict)
```

```python
#slope
model.coef_
```

```python
#intercept
model.intercept_
```

```python
model.predict([[5.6]])
```

```python
model.predict([[9.6]])
```

```python
#slope formula --> other way to check output
Y=0.55*X-0.911
```

```python
Y=0.55*(7.8)-0.911
```

```python
Y
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,model.predict(X_train),color="red")
```
