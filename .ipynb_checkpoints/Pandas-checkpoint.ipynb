```python
import pandas as pd 
employees_data = {
    'id':[1,7,11,90,3],
    'name':['Alice','Bob','Meir','Winton','Jonathan']
}

employees_uni_data = {
    'id' : [3,11,90],
     'unique_id': [1,2,3]
}
employee_df = pd.DataFrame(employees_data)
employee_uni_df = pd.DataFrame(employees_uni_data)
```

```python
result_df = pd.merge(employee_df,employee_uni_df, on = 'id', how='left')
result_df
```

```python
result_df = pd.merge(employee_df,employee_uni_df, on = 'id', how='right')
result_df
```

```python
result_df = pd.merge(employee_df,employee_uni_df, on = 'id', how='inner')
result_df
```

```python
result_df = pd.merge(employee_df,employee_uni_df, how='cross')
result_df
```

```python

result_df = pd.merge(employee_df,employee_uni_df, on = 'id', how='outer')
result_df
```

```python
result_df = pd.concat([employee_df,employee_uni_df], ignore_index =True)
result_df
```

```python
# to represent column wise data use axis=1
```

```python
result_df = pd.concat([employee_df,employee_uni_df], axis =1)
result_df
```

```python
def convert(co):
    if co == "BMW X5":
        return "Maruti Suzuki-800"
    else:
        return co
```

```python
import pandas as pd
df = pd.DataFrame({
    'A':[1,-999,3,-999],
    'B':['x','y','z','x']
})
print(df)
```

```python
df2 =pd.DataFrame({
    'Grade':['A','F','B','F','C'],
    'Score':[90,-1,85,-1,75]
})
print(df2)
```

```python
replaced_dict = {'F':'fail',-1:0}
df2_replaced = df2.replace(to_replace =replaced_dict)
print(df2_replaced)
```
