```python
sample_mean_iq = 110
mean_population =100
population_std = 15
```

```python
data=population_std * randn(50)+sample_mean_iq
```

```python
np.mean(data) 
```

```python
np.std(data)
```

```python
null_hypothesis = mean_population
```

```python
pip install statsmodels
```

```python
from statsmodels.stats.weightstats import ztest
# assuming 'data' is your sample data (like a numpy array or a pandas series)
# and 'null hypothesis' is the value you want to test against ( like 0 or a known mean)
z_stat, p_value = ztest(data, value = null_hypothesis, alternative = 'larger')
print(f"z-statistic: {z_stat}")
print(f"p-value: {p_value}")
# alternative 'larger' means you are testing if the
#sample mean is significantly greater tham the null hypothesis mean
```

```python
if p_value<alpha:
    print("reject the null hypothesis")
else:
    print("fail to reject the null hypothesis")
```

# chi-square-Categorical data

```python
dataset = sns.load_dataset('tips')
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
dataset[["sex","smoker"]]
```

```python
import pandas as pd
dataset_table = pd.crosstab(dataset["sex"], dataset["smoker"])
```

```python
dataset_table
```

```python
dataset_array=dataset_table.values
```

```python
observed_value =dataset_array
```

```python
import scipy.stats as stats
```

```python
stats_test, p, dof, Expected_Value = stats.chi2_contingency(dataset_array)
```

```python
Expected_Value
```

```python
no_of_col =dataset_table.iloc[0:2,0]
```

```python
len(no_of_col)
```

```python
no_of_row =dataset_table.iloc[0,0:2]
```

```python
len(no_of_row)
```

```python
dof= (len(no_of_col)-1)*(len(no_of_row)-1)
```

```python
dof
```

```python
for o,e in zip(observed_value,Expected_Value):
    print(o,e)
```

```python
from scipy.stats import chi2
chisqure = sum([(o-e)**2/e for o,e in zip(observed_value,Expected_Value)])
```

```python
chisqurestats=chisqure[0]+chisqure[1]
```

```python
chisqurestats
```

```python
alpha = 0.05
CI = 0.95
```

```python
critical_chi_value = chi2.ppf(CI, df = dof)
```

```python
critical_chi_value
```

```python
if chisqurestats<critical_chi_value:
    print("Reject the null hypothesis- there is relationship between two variable")
else:
    print("accept the null hypothesis- there is no relationship between two variable")
    
```
