```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.arange(0,10)
y= np.arange(11,21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Gragh in 2D')
plt.savefig('test.png')
```

```python
y = x*x
```

```python
y
```

```python
# line plot
plt.plot(x,y, 'r+',linestyle = 'dashed',linewidth=2,markersize =12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d diagram')
```

```python
#creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
```

```python
x = np.arange(0,20)
y = np.arange(2,20)
```

```python
y = x*x*x*x
```

```python
y
```

```python
#creating subplots
plt.subplot(2,4,1)
plt.plot(x,y,'r--')
plt.subplot(2,4,2)
plt.plot(y,x,'g--')
plt.subplot(2,4,3)
plt.plot(x,y,'b--')
plt.subplot(2,4,4)
plt.plot(x,y,'b*')
plt.subplot(2,4,5)
plt.plot(x,y,'y*')
plt.subplot(2,4,6)
plt.plot(y,x,'y--')
plt.subplot(2,4,7)
plt.plot(x,y,'b.')
plt.subplot(2,4,8)
plt.plot(x,y,'b*')
```

```python
#histogram
a = np.array([22,87,5,43,73,55,67,88])
plt.hist(a,bins = 4)
```

```python
#data to plot
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0,0) #explode 1st slice

#plot
plt.pie(sizes, explode = explode, labels = labels, colors=colors, autopct = '%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
