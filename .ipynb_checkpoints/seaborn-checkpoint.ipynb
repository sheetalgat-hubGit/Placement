```python
import seaborn as sns
import matplotlib.pyplot as plt

#load sample  dataset 
data = pd.DataFrame({
    'maths':[85, 90,78,95,88],
    'Science':[88,85,80,92,90],
    'English':[75,80,70,85,82]
})
data
```

```python
#calculate correlation matrix
#it shows correlation with subjects
corr = data.corr()

#heatmap
#annnot means it shows numbers on plot
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Between Subjects")
plt.show()
```

```python
#for specific column correlation done
corr= data[['maths','Science']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Between Subjects")
plt.show()
```

```python
df=sns.load_dataset('tips')
df
```

```python
corr1= df[['total_bill','tip','size']].corr()
sns.heatmap(corr1,annot=True,cmap='coolwarm')
plt.title("Correlation Between features")
plt.show()
```

```python
#used to check on whethers
#KDEplot
rainfall =[100,120,130,110,150,140,160]
#KDEplot
sns.kdeplot(rainfall, shade = True)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall (mm)")
plt.show()

```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.kdeplot(df[['total_bill']], shade = True)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall (mm)")
plt.show()
```

```python
runs =[45,67,23,89,12,56,78]
sns.histplot(runs,kde = True)
plt.title("Distribution of Runs")
plt.xlabel("Runs scored")
plt.show()
```

```python
data = pd.DataFrame({
    'Height':[150,165,170,155,180],
    'Weight':[50,60,65,55,70]
})
sns.scatterplot(x='Height',y='Weight', data = data)
plt.title("Height vs Weight")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip', data = df)
plt.title("Total_bill vs Tip")
plt.show()
```

```python
days = [1,2,3,4,5]
temp = [28,30,32,29,31]

sns.lineplot(x=days, y = temp)
plt.title("Temperature over 5 days")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
#to check outliers
data= pd.DataFrame({
    'Region':['North','South','North','South','East'],
    'Rainfall':[100,150,120,200,130]
})
sns.boxplot(x= 'Region',y ='Rainfall',data=data)
plt.title("Rainfall by Region")
plt.show()
```
